## 3. Feature Enginering

```python
# a. BMI Category

def bmi_category(bmi):
    if bmi < 18.5:
        return "underweight"
    elif bmi < 25:
        return "normal"
    elif bmi < 30:
        return "overweight"
    else:
        return "obese"

df["bmi_category"] = df["bmi"].apply(bmi_category)
```

```python
# b. BP Category (Blood Pressure)

def bp_category(row):
    sbp = row["systolic_bp"]
    dbp = row["diastolic_bp"]

    if sbp < 120 and dbp < 80:
        return "Normal"
    elif 120 <= sbp < 130 and dbp < 80:
        return "Elevated"
    elif 130 <= sbp < 140 or 80 <= dbp < 90:
        return "Hypertension Stage1"
    else:
        return "Hypertension Stage2"

df["bp_category"] = df.apply(bp_category, axis=1)
```

```python
# c. Cholesterol Ratio (total / HDL)

df["cholesterol_ratio"] = df["cholesterol_total"] / df["hdl_cholesterol"]

# Risiko tinggi jika > 5
df["cholesterol_ratio_flag"] = (df["cholesterol_ratio"] > 5).astype(int)
```

```python
# d. Age Group

df["age_group"] = pd.cut(
    df["age"],
    bins=[0, 30, 45, 60, 120],
    labels=["Young", "Adult", "Middle Age", "Elderly"]
)
```

```python
df["lifestyle_score"] = (
    (df["physical_activity_minutes_per_week"] >= 150).astype(int) +
    (df["diet_score"] >= df["diet_score"].median()).astype(int) +
    (df["sleep_hours_per_day"] >= 7).astype(int) -
    (df["alcohol_consumption_per_week"] > 7).astype(int) -
    (df["smoking_status"].isin(["current", "former"])).astype(int)
)
```

## 4. Deteksi Data Outlier

```python
# a. Mendeteksi Outlier (Menggunakan IQR)

Q1 = df[num].quantile(0.25)
Q3 = df[num].quantile(0.75)
IQR = Q3 - Q1

# Filter Baris Outlier
outlier_iqr = df[((df[num] < (Q1 - 1.5 * IQR)) | (df[num] > (Q3 + 1.5 * IQR))).any(axis = 1)]
print("Data Outlier :", outlier_iqr.shape[0], "Baris")
```

```python
df.shape
```

Terdapat Data Outlier Sebanyak 35599 Baris

```python
# b. Melihat Outlier di Kolom Apa Saja

outlier_count = {}

for num in df.select_dtypes(include='number').columns:
    Q1 = df[num].quantile(0.25)
    Q3 = df[num].quantile(0.75)
    IQR = Q3 - Q1

    # hitung outlier untuk kolom tersebut
    outliers = df[(df[num] < (Q1 - 1.5 * IQR)) | (df[num] > (Q3 + 1.5 * IQR))]
    outlier_count[num] = outliers.shape[0]

# tampilkan hasil
for col, count in outlier_count.items():
    print(f"Jumlah outlier Kolom {col}: {count}")
```
