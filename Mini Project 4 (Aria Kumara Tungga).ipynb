## 3. Deteksi Outlier

```python
# a. Mendeteksi Outlier (Menggunakan IQR)
Q1 = df[numerik].quantile(0.25)
Q3 = df[numerik].quantile(0.75)
IQR = Q3 - Q1

# Filter Baris Outlier
outlier_iqr = df[((df[numerik] < (Q1 - 1.5 * IQR)) | (df[numerik] > (Q3 + 1.5 * IQR))).any(axis = 1)]
print("Data Outlier :", outlier_iqr.shape[0], "Baris")
```

Data Outlier Bernilai 0 Baris / Tidak ada Outlier
